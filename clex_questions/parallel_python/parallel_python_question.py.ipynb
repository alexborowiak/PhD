```python
import xarray as xr
import os, sys
```

```python
ROOT_DIR = '/g/data/w40/ab2313/PhD/longrunmip/tas/regrid_retimestamped'
```

```python
fnames = os.listdir(ROOT_DIR)
len(fnames)
```

```python
def open_and_save_netcdf(fname, ROOT_DIR):
    fpath = os.path.join(ROOT_DIR, fname)
    print(f'Opening {fpath}')
    ds = xr.open_dataset(fpath, use_cftime=True)
    ds = ds.isel(time=0, lat=0, lon=0)
    ds.to_netcdf(f'{fname}.nc', )
```

```python
for fname in fnames:
    open_and_save_netcdf
```
