```python
import subprocess


def execute_shell_cmd(cdm):
    
    process = subprocess.Popen(cdm, universal_newlines=True, shell=True, executable='/bin/bash',
                              stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    stdout, stderr = process.communicate()
    if stderr:
        print(stderr)
    return stdout
```
