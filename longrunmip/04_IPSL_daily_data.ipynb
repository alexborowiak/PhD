```python
ROOT_DIR1 = '/g/data/oi10/replicas/CMIP6/ScenarioMIP/IPSL/IPSL-CM6A-LR/ssp585/r1i1p1f1/Amon/tas/gr/v20190903/'
ROOT_DIR0 = '/g/data/oi10/replicas/CMIP6/CMIP/IPSL/IPSL-CM6A-LR/historical/r1i1p1f1/Amon/tas/gr/v20180803/'
```

```python
d0 = xr.open_dataset(ROOT_DIR0 + 'tas_Amon_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001-201412.nc')
```

```python
d1 = xr.open_mfdataset(ROOT_DIR1 + "*.nc")
```

```python
d = d1.combine_first(d0)
```

```python
d.time.values
```

## Creating anomalies

```python
climatology = d0.where(d0.time.dt.year.isin(np.arange(1850,1901)), drop = True).mean(dim = ['lat','lon','time'])
```

```python
d_mean = d.mean(dim = ['lat','lon']) - climatology
```

```python
d_mean
```

```python
d_mean_roll = d_mean.rolling(time = 30 * 12).mean().compute()
```

```python
d_mean.tas.plot()
d_mean_roll.tas.plot()
```
