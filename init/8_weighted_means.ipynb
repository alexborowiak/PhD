# Weighted Means

This notebook pertains to exploring the difference between normal and area weighted means. Code is from
http://xarray.pydata.org/en/stable/examples/area_weighted_temperature.html

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = xr.tutorial.load_dataset("air_temperature")

# to celsius
air = ds.air - 273.15

# resample from 6-hourly to daily values
air = air.resample(time="D").mean()

air
```

```python
weights = np.cos(np.deg2rad(air.lat))
weights.name = "weights"
weights
```

```python
air.weighted(weights).mean(('lon','lat'))
```

```python
air_weighted = air.weighted(weights)
air_weighted
```

```python
weighted_mean = air_weighted.mean(("lon", "lat"))
weighted_mean
```

```python
weighted_mean.plot(label="weighted")
air.mean(("lon", "lat")).plot(label="unweighted")

plt.legend()
```
