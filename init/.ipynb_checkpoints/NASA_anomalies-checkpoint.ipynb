```python
abrupt4 = xr.open_mfdataset(ROOT_DIR1 + '*.nc')
hist = xr.open_mfdataset(ROOT_DIR2 + '*.nc')
```

```python
abrupt4
```

```python
hist
```

```python
hist_clima = hist.where(hist.time.dt.year.isin(np.arange(1850,1900)))
hist_clima
```

```python
hist_clima_mean = hist_clima.ts.groupby('time.month').mean(dim = 'time')
hist_clima_mean 
```

```python
 (hist_clima_mean - 273).plot(x = 'lon', y ='lat', col = 'month', col_wrap = 4)
```

```python
clim_stor = []
for month in (hist_clima_mean.month.values):
    sub_month = abrupt4.where(abrupt4.time.dt.month.isin([month]), drop = True)
    clim_month = hist_clima_mean.sel(month = month)
    
    anom_month = sub_month.ts - clim_month
    clim_stor.append(anom_month)
    

abrupt4_anom = xr.concat(clim_stor, dim = 'time').drop('month')
```
