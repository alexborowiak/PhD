# Functions

```python
def climatology(hist, start = 1850, end = 1901):

    ### CLIMATOLOGY

    # Getting just the years for climatology. This should be for each pixel, the mean temperature
    # from 1850 to 1900. 
    climatology = hist.where(hist.time.dt.year.isin(np.arange(start,end)), drop = True)\
                        .mean(dim = 'time')

    return climatology


def anomalies(data, hist):

    climatology = climatology(hist)

    data_resampled = data.resample(time = 'Y').mean()
    data_anom = (data_resampled - climatology).chunk({'time':8})


    return data_anom

def space_mean(data):
    
        # Lat weights
        weights = np.cos(np.deg2rad(data.lat))
        weights.name= 'weights'
        
        # Calculating the weighted mean.
        data_wmean = data.weighted(weights).mean(dim = ['lat','lon'])

        return data_wmean
```

```python
WINDOW_LENGTH = 120
WINDOW_DIFF = 0
```

```python
len(ech1pct_smean.time.values)
```

```python
DIST1_START = 0
DIST1_END = WINDOW_LENGTH
DIST2_START = WINDOW_LENGTH + WINDOW_DIFF
DIST2_END = 2 * WINDOW_LENGTH + WINDOW_DIFF
```

```python
effective_length = len(ech1pct_smean.time.values) - DIST2_END
effective_length
```

```python
ks_pval_array = np.zeros(effective_length)
ks_pval_array.shape
```

```python
for t in np.arange(effective_length):
    p1 = ech1pct_smean.isel(time = slice(t + DIST1_START, t + DIST1_END))
    p2 = ech1pct_smean.isel(time = 
                            slice(t + DIST2_START,  t + DIST2_END))
    
    ks_pval = kstest(p1,p2)[1]
    
    ks_pval_array[t] += ks_pval
```

```python
ks_pval_array
```

```python
time = ech1pct_smean.time.values[:effective_length]
```

```python
ks_pval_ds = xr.DataArray(ks_pval_array,
                          coords = [time], dims = ['time']) #+ pd.to_timedelta(f'{WINDOW_LENGTH/2}Y')],
ks_pval_ds 
```

```python
fig = plt.figure(figsize = (15,8))

ax = fig.add_subplot()
time = ech1pct_smean.time.values[:effective_length]

ax.plot(ks_pval_ds.time.values,ks_pval_ds.values)

ax.plot([time[0], time[-1]], [0.05, 0.05], linestyle = ':', linewidth = 2)
ax.plot([time[0], time[-1]], [0.1, 0.1], linestyle = ':', linewidth = 2)
```
